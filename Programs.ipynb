```python
# 1.Write a function to add two numbers and it should return the sum
def add(a,b):
  return a+b
add(10,20)
```

```python
# 2. Write a function to calculate the factorial of a number
def fact(n):
  if n==1:
    return 1
  else:
    return n*fact(n-1)
fact(5)
```

```python
# 3. Create a class the class name is student with name,Rollno,marks and add a method to display student details
class Student:
  def __init__(self,name,rollno,marks):
    self.name=name
    self.rollno=rollno
    self.marks=marks
  def display(self):
    print(self.name)
    print(self.rollno)
    print(self.marks)
s=Student("Namratha",123,90)
s.display()
```

```python
# 4. Create a class the class name is rectangle with length and width are attributes and methods are calculate area and calculate the perimeter
class Rectangle:
  def __init__(self,length,width):
    self.length=length
    self.width=width
  def calculate_area(self):
    area= self.length*self.width
    return area
  def calculate_perimeter(self):
    perimeter= 2*(self.length+self.width)
    return perimeter
r=Rectangle(10,20)
print(r.calculate_area())
print(r.calculate_perimeter())
```

```python
# 5. Create a class with the class name of bank account with accno and balance along with methods are deposit,withdraw and display balance
class bank_account:
  def __init__(self,accno,balance):
    self.accno=accno
    self.balance=balance
  def deposit(self,amount):
    self.balance+=amount
    print("deposited",amount,"new balance is",self.balance)
  def withdraw(self,amount):
    if amount>self.balance:
      print("insufficient balance")
    else:
      self.balance-=amount
      print("withdrawn",amount,"new balance is",self.balance)
  def display(self):
    print("Accno:",self.accno)
    print("Balance:",self.balance)
b=bank_account(123456789,10000)
b.deposit(5000)
b.withdraw(2000)
b.display()
```
