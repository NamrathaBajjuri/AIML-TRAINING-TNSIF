```python
# Student Report Card Program
import numpy as np
import pandas as pd
```

```python
# List to store student names
student_names = []
```

```python
# Taking student details
name = input("Enter student name: ")
age = int(input("Enter student age: "))
```

```python
# Store name in list
student_names.append(name)
```

```python
# Tuple to store student information
student_info = (name, age)
```

```python
# Set to store subjects
subjects = set()
```

```python
# Dictionary to store marks
marks = {}
```

```python
# Number of subjects
num_subjects = int(input("Enter number of subjects: "))
```

```python
# Taking subjects and marks
for i in range(num_subjects):
    subject = input("Enter subject name: ")
    mark = int(input("Enter marks: "))
    # Store subject in set
    subjects.add(subject)
    # Store marks in dictionary
    marks[subject] = mark
```

```python
# Calculate total marks
total_marks = sum(marks.values())
```

```python
# Calculate average marks
average_marks = total_marks / num_subjects
```

```python
# Printing Report Card

print("\n----- Student Report Card -----")
print("Name:", student_info[0])
print("Age:", student_info[1])
print("Subjects:", subjects)
print("Marks:")
for subject, mark in marks.items():
    print(subject, ":", mark)
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
```
