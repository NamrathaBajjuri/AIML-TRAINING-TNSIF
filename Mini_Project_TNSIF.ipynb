```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
content=df['type'].value_counts()
print(content)
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title("Netflix Content Types")
plt.xlabel("Content Types")
plt.ylabel("Count")
plt.show()
```

```python
print(df['country'].value_counts())
```

```python
df['country'].value_counts().head(10).plot(kind='pie')
plt.title("Countries contribute most content in Netflix")
plt.show()
```

```python
director=df['director'].value_counts().head(10)
print(director)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=director.index,y=director.values)
plt.title("Top 10 Directors with the most content in Netflix")
plt.xlabel("Directors")
plt.ylabel("Directors and Titles")
plt.xticks(rotation=45,ha='right')
plt.show()
```
