# **PYTHON PROGRAM USING NUMPY**

```python
import numpy as np
# Create NumPy array
marks = np.array([65, 78, 90, 55, 88])
# Find average marks
average = np.mean(marks)
# Find highest marks
highest = np.max(marks)
# Find lowest marks
lowest = np.min(marks)
# Display results
print("Marks:", marks)
print("Average Marks:", average)
print("Highest Marks:", highest)
print("Lowest Marks:", lowest)
```
