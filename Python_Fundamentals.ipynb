# **OPERATORS**

```python
# Program on Arthmetic operators
a=20
b=10
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
```

```python
# Program on Comparision operators
a=20
b=10
print(a<b)
print(a>b)
print(a>=b)
print(a<=b)
print(a==b)
print(a!=b)
```

```python
# Program on Logical operators
a=True
b=False
print(a and b)
print(a or b)
print(not b)
```

```python
# Program on Assignment operators
x=2
x+=1
print(x)
x-=1
print(x)
x*=2
print(x)
x/=2
print(x)
x%=2
print(x)
```

# **DATA TYPES**

```python
# Program on List

colors=['red','yellow','blue']
print(colors)
colors.append('green')
print(colors)
colors.insert(2,'orange')
print(colors)
colors.remove('green')
print(colors)
colors.pop(2)
print(colors)
colors.sort()
print(colors)
colors.reverse()
print(colors)
```

```python
# Program on Tuple
car_brands=("benz","suzuki","thar","thar","bmw")
print(car_brands)
print(len(car_brands))
print(car_brands.count('thar'))
print(car_brands[2])
```

```python
# Program on Sets

s1={1,2,3,4,5}
s2={4,5,6,7,8}
print(s1)
print(s2)
s1.add(6)
print(s1)
s1.remove(3)
print(s1)
unions=s1.union(s2)
print(unions)
inter=s1.intersection(s2)
print(inter)
diff=s1.difference(s2)
print(diff)
s1.clear()
print(s1)
```

```python
# Program on Dictionary

stud={"name":"namratha","age":21,"course":"aiml"}
print(stud)
print(stud.keys())
print(stud.values())
print(stud.items())
print(stud.get("name"))
stud.pop("name")
print(stud)
stud.update({"name":"namratha"})
print(stud)
stud.update({"age":20})
print(stud)
```
