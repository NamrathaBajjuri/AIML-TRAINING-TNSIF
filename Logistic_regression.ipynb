# **LOGISTIC REGRESSION**

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
# Step 2: Create Dataset
df = pd.DataFrame({
    "Loan":[5,10,20,70,80,90],
    "Status":[1,1,1,0,0,0]
})

print("Dataset")
print(df)
```

```python
# Step 3: Input and Output
X = df[["Loan"]]
y = df["Status"]
```

```python
# Step 4: Create Model
model = LogisticRegression()
```

```python
# Step 5: Train Model
model.fit(X,y)
```

```python
# Step 6: Prediction
prediction = model.predict([[60]])
```

```python
# Step 7: Output
print("\nPrediction:",prediction[0])

if prediction[0]==1:
    print("Loan Approved")
else:
    print("Loan Rejected")
```

```python
# Scatter Plot
plt.scatter(df["Loan"], df["Status"], color="blue")

plt.title("Loan vs Status")
plt.xlabel("Loan")
plt.ylabel("Status")

plt.show()
```

```python
# Count Plot
sns.countplot(x="Status", data=df)

plt.title("Loan vs Status")
plt.xlabel("Status")
```
