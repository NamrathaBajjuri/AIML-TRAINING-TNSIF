# Markov Decision Process (MDP):

Markov Decision Process is a mathematical framework used in Reinforcement Learning to represent decision-making problems.

It contains:

State (S): Current situation

Action (A): Possible choices

Reward (R): Feedback

Transition (P): Movement from one state to another

Policy (π): Strategy used to select actions

# Example:

In a game:

State → Player position

Action → Move left/right

Reward → Score gained

```python
states = ["Start", "Middle", "Goal"]
actions = ["Move"]

for state in states:
    print("Current State:", state)
    print("Action:", actions[0])
```

# Output:

Displays states and actions.
