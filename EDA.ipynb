```python
df=pd.read_csv("weight-height.csv")
df
```

```python
df.describe()
```

```python
#Detect outliers using IQR
Q1 = df.Height.quantile(0.25)
Q3 = df.Height.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
#Here are the outliers

df[(df.Height<lower_limit)|(df.Height>upper_limit)]
```

```python
# Calculate mean
mean_value = df['Height'].mean()

# Replace outliers with mean
df['Height'] = df['Height'].apply(
    lambda x: mean_value if x < lower_limit or x > upper_limit else x
)
# Print dataset
print(df)
```

```python
# Calculate median
median_value = df['Height'].median()

# Replace outliers with median
df['Height'] = df['Height'].apply(
    lambda x: median_value if x < lower_limit or x > upper_limit else x
)

# Print dataset
print(df)
```
