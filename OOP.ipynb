# **CLASS**

```python
class person:
  name="Namratha"
p=person()
q=person()
print(p.name)
print(q.name)
```

# **OBJECT**

```python
class stud:
  sid=102
  sname="Namratha"
obj=stud()
print("The Student ID is:", obj.sid, "and the Student Name is:", obj.sname)
```

# **METHOD**

```python
class employee:
  def greet(self):
    print("Hello")
e1=employee()
e1.greet()
```

```python
class student:
    # Constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age
s1=student("Namratha",21)
print("Name : ",s1.name)
print("Age : ",s1.age)
```
