# **PYTHON PROGRAM USING** **PANDAS**

```python
# Import pandas library
import pandas as pd
# Create dataset using dictionary
data = {
    "Student_ID": [101, 102, 103, 104, 105],
    "Name": ["Namratha", "Ravi", "Sita", "John", "Priya"],
    "Marks": [85, 78, 92, 65, 88],
    "Department": ["CSE", "ECE", "CSE", "IT", "EEE"]
}
# Create DataFrame
df = pd.DataFrame(data)
# Display complete dataset
print("Student Data:")
print(df)
# Display first rows
print("\nFirst 5 rows:")
print(df.head())
# Display information about dataset
print("\nDataset Information:")
print(df.info())
# Display statistical summary
print("\nStatistical Summary:")
print(df.describe())
# Find average marks
print("\nAverage Marks:")
print(df["Marks"].mean())
# Find highest marks
print("\nHighest Marks:")
print(df["Marks"].max())
# Find lowest marks
print("\nLowest Marks:")
print(df["Marks"].min())
# Filter students with marks above 80
print("\nStudents scoring above 80:")
print(df[df["Marks"] > 80])
# Sort data by marks
print("\nSorted Marks:")
print(df.sort_values("Marks"))
```
