# **Reinforecement Learning**
# 1. What is Reinforcement Learning?

Reinforcement Learning is a branch of Machine Learning where an agent learns by interacting with an environment. Instead of learning from labeled data, the agent receives rewards or penalties based on its actions and gradually learns the best strategy to maximize the total reward.

# 2. Real-Life Applications

Self-driving cars

Robotics

Game AI (Chess, Go, Atari)

Recommendation Systems

Traffic Signal Optimization

Stock Trading

Resource Management

# 3. Components of Reinforcement Learning


*   Agent : Learner that makes decisions
*   Environment : World where the agent acts


*   State(S) : Current situation
*   Action(A) : Decision taken by agent


*   Reward(R) : Feedback received

# Workflow of Reinforcement Learning
         +------------------+
        |     Environment  |
        +------------------+
              ↑       |
          Reward    State
              |       ↓
        +------------------+
        |      Agent       |
        +------------------+
              |
            Action
              |
              ↓
        Environment

# Advantages
*  Learns through experience
*   No labeled dataset required


*   Solves sequential decision problems
*   Adapts to changing environments

*   Used in robotics, gaming, and automation


# Disadvantages
*   Requires many training episodes
*   Computationally expensive

*   Slow convergence
*   Sensitive to reward design

*   Difficult for very large state spaces

# Example:
A robot learns to move by receiving rewards for correct movements.

```python
import random

actions = ["Left", "Right"]

for i in range(5):
    action = random.choice(actions)
    reward = random.randint(0,10)

    print("Action:", action)
    print("Reward:", reward)
```

```python
# Bar Chart
import matplotlib.pyplot as plt
reward_list = [10, 20, 30, 40, 50]
plt.figure(figsize=(6,4))
plt.bar(range(1, 6), reward_list)
plt.title("Reward Received in Each Step")
plt.xlabel("Step")
plt.ylabel("Reward")
plt.xticks(range(1, 6))
plt.show()
```
