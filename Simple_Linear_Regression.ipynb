# **SIMPLE LINEAR REGRESSION**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Dataset
df = pd.DataFrame({
    "House_Size":[500,700,900,1100,1300,1500,1700,2100,2300],
    "House_Prize":[25,35,45,55,65,75,85,105,115]
})

print(df)
```

```python
# Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(df["House_Size"], df["House_Prize"], color='blue')
plt.title("House_Size vs House_Prize")
plt.xlabel("House_Size")
plt.ylabel("House_Prize")
plt.grid(True)
plt.show()
```

```python
sns.countplot(x="House_Prize", data=df)

plt.title("House_Size vs House_Prize")
plt.xlabel("House_Size")
plt.ylabel("House_Prize")
plt.show()
```

```python
# Independent Variable
X = df[["House_Size"]]

# Dependent Variable
y = df["House_Prize"]

# Create Model
model = LinearRegression()

# Train Model
model.fit(X,y)
```

```python
# Prediction
prediction = model.predict([[1250]])

print("Predicted House Prize:",prediction[0])

print("Slope:",model.coef_[0])
print("Intercept:",model.intercept_)
```
