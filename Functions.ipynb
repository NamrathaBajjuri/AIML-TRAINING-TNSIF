# **FUNCTIONS**

```python
# Basic Function Program
def function_name():
  #function body
  print("HELLO WORLD!")
function_name()
```

```python
# Program on Function Without Parameters and Without Return Value
def display():
  print("Learning Python")
display()
```

```python
# Program on Function With Parameters
def greet(name):
  print("Hii",name)
greet("Namratha")
greet("Karthik")
```

```python
# Program on Function With Return Value
def add(a,b):
  return a+b
result=add(10,20)
print(result)
```

```python
# Program on Function With Parameters and Return Value
def square(num):
  return num*num
result=square(5)
print(result)
```
