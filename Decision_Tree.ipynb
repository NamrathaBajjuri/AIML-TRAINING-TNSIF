# **DECISION TREE**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier

# Dataset
df = pd.DataFrame({
    "Study_Hours": [2, 4, 6, 8, 3, 5, 7, 9, 1, 10],
    "Attendance": [60, 70, 80, 90, 65, 75, 85, 95, 50, 98],
    "Assignments": [1, 2, 3, 4, 1, 2, 3, 4, 0, 5],
    "Result": ["Fail", "Fail", "Pass", "Pass", "Fail","Pass", "Pass", "Pass", "Fail", "Pass"]
})

print(df)
```

```python
#Count Plot Data visualization
sns.countplot(x="Result", data=df)

plt.title("Pass vs Fail")
plt.xlabel("Result")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(6,4))

sns.scatterplot(x="Study_Hours",
                y="Attendance",
                hue="Result",
                data=df,
                s=100)

plt.title("Study Hours vs Attendance")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6,4))

sns.boxplot(x="Result",
            y="Study_Hours",
            data=df)

plt.title("Study Hours by Result")

plt.show()
```

```python
# Independent Variable
X = df[["Study_Hours","Attendance","Assignments"]]

# Dependent Variable
y = df["Result"]

# Create Model
model = DecisionTreeClassifier()

# Train Model
model.fit(X,y)
```

```python
# Prediction
prediction = model.predict([[7.5,70,4]])

print("Predicted Result:",prediction[0])
```
