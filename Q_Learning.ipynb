# Q-Learning Algorithm:

Q-Learning is a model-free Reinforcement Learning algorithm that helps an agent learn the best action by calculating Q-values.

Q-value represents:

The expected reward for taking an action in a particular state.

Formula:

Q(s,a) = Q(s,a) + α[R + γ max Q(s',a') - Q(s,a)]

Where:

α = Learning rate

γ = Discount factor

R = Reward

```python
import numpy as np

states = 5
actions = 2

Q = np.zeros((states, actions))

reward = 10
learning_rate = 0.8
discount = 0.9

Q[0][1] = Q[0][1] + learning_rate * (
    reward + discount*np.max(Q[1]) - Q[0][1]
)

print(Q)
```
