# Anomaly Detection


Anomaly Detection is a technique used to identify unusual or abnormal patterns in data that are different from normal behavior.

# Examples:

*  Fraud detection in banking
*  Detecting network attacks in cybersecurity

*  Finding machine faults in industries

# Types:
# Supervised Anomaly Detection:
Uses labeled data (normal and abnormal).
# Unsupervised Anomaly Detection:
Finds unusual patterns without labels.
# Semi-supervised Anomaly Detection:
Uses mostly normal data with few abnormal examples.

```python
import numpy as np
from sklearn.ensemble import IsolationForest

# Sample data
data = np.array([[10], [12], [11], [13], [100], [9], [14]])

# Model
model = IsolationForest(contamination=0.2)

# Training
model.fit(data)

# Prediction
prediction = model.predict(data)

print(prediction)
```

# Explanation
1 → Normal data

-1 → Anomaly detected
